```python
#Zero shot approach
import pandas as pd
from transformers import pipeline
import warnings
warnings.filterwarnings("ignore")


data1 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2026\Kapture_Reports\Outputs\PL_Outputs.xlsx"
df= pd.read_excel(data1)
df['ssi'] = df['Description']

classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")

print(['ssi'].head(10))

my_categories = [
    "Technician No Show or Fake Update or data mismatch",
    "Replacement Inquiry",
    "Refund Inquiry",
    "SLA Breach and Delay",
    "General claim status Inquiry",
    "Not in scope and name mismatch",
    "Escalation and SLA Breaches"
]

def ai_classifier(text):
    clean_text = str(text).lower().replace('undefined - undefined - ', '')
    result = classifier(clean_text, my_categories)
    return result['labels'][0]

print("Categorizing tickets...")
df['ssi'] = df['Description'].apply(ai_classifier)

pd.set_option('display.max_colwidth', None)
print(df[['Description', 'ssi']])
```
