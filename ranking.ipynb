```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data1= r"C:\Users\rajeshkumar.t\Desktop\XX\Ranking.csv"
df= pd.read_csv(data1)
```

```python
print(df.head(3))
df= df.dropna(subset=['LOB'])
```

```python
print(df.head(3))
```

```python
lob_issue_counts = df.groupby(['LOB','issue_type'], as_index=False)['contacts'].sum().reset_index()
lob_issue_counts['rank'] = lob_issue_counts.groupby('LOB')['contacts'].rank(ascending=False, method='first')
top_3_per_lob = lob_issue_counts[lob_issue_counts['rank']<=3].sort_values(['LOB', 'rank'])
print("Top 3 issues per LOB")
print(top_3_per_lob.head(15))
```
