```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data4= r"C:\Users\rajeshkumar.t\Desktop\ML\Cohort_1.csv"
df = pd.read_csv(data4)
print(df)
```

```python
df['churn_rate'] = (df['active_users'] - df['user_next_month'])/ df['user_next_month']
churn_pivot = df.pivot(index = 'cohort_month', columns= 'month_index', values= 'churn_rate')
print(df)
```

```python
#df['churn_pivot'] =df['churn_pivot'].fillna(0)
plt.figure(figsize=(10,6))
sns.heatmap(churn_pivot, annot=True, fmt= '.1%', cmap= 'Reds')
plt.title('Churn Rate Heatmap')
plt.show()

sns.scatterplot(data=df, x= 'active_users', y='churn_rate')
plt.title("Impact of delivery Breach")
plt.show()

           
```
