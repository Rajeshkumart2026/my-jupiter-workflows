```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data = r"G:\My Drive\New_Working_Files\KNIME\CTU_2026\ML_Models\Order_raw_data.csv"
df = pd.read_csv(data)

context_data = {
    "product": df['product_title'].iloc[0],
    "trade_in_brand": df['prexo_incoming_brand'].iloc[0],
    "customer_reason": df['prexo_reason'].iloc[0],
    "exchange_value": df['adjustment_amount_for_prexo'].iloc[0],
    "listing_price": df['listing_price'].iloc[0]
        }

prompt = f"""
SYSTEM: You are a E-commerce Fraud Detection AI.
DATA:
- Item Brought: {context_data['product']}
- Trade-in Item: {context_data['trade_in_brand']}
- Reported Condition: {context_data['customer_reason']}
- Price Adjsutment Reason: {context_data['exchange_value']}

TASK:
Does it make sense to give a {context_data['exchange_value']} discount
for a trade-in with condition '{context_data['customer_reason']}'?
If the adjustment is higher then 50% of the listing price, flag it.
"""
ai_response_text = f"The condition is {context_data['customer_reason']}. Flag it as suspicious due to missing data."

df['is_suspicious'] = (df['prexo_reason'].isna()) & (df['adjustment_amount_for_prexo']>0)

def fraud_safety_value(ai_output, order_status):
    if "flag it" in ai_output.lower() or "nan" in ai_output.lower():
        new_status = "UNDER_REVIEW"
        notification = "Order put in hold. Missing trade in verification for discount."
        return new_status, notification
    return order_status, "Order is clean."

status, msg = fraud_safety_value(ai_response_text, df['status'].iloc[0])
print(f"Automation Result: Status changed to {status}. Alert: {msg}")
    
print("-----AI input Prompt-----")
print(prompt)
print("--------------")
print(f"Automation Result: status changed to {status}. Alert: {msg}")
```
