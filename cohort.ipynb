```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data3 = r"C:\Users\rajeshkumar.t\Desktop\ML\Cohort.csv"
df= pd.read_csv(data3)
print(df.columns)
```

```python
cohort_pivot = df.pivot(index='cohort_month', columns ='month_number', values='active_users')
cohort_size= cohort_pivot.iloc[:,0]
retention_matrix = cohort_pivot.divide(cohort_size, axis=0) 
```

```python
print(retention_matrix)
print(cohort_pivot)

plt.figure(figsize=(10,6))
plt.title("Monthly retention pct")
sns.heatmap(retention_matrix,
            annot=True,
            fmt='.1%',
            cmap='Reds',
            linewidths =.5
           )
plt.ylable = ("Cohort of the first value")
plt.xlable = ("Months of purchase")
plt.show()
```
