```python
df_daily = df_daily.sort_values('session_date')
df_daily['seven_days_mavg'] = df_daily['total_calls'].rolling(window=7).mean()
df_daily['trend'] = range(len(df_daily))
baseline= df_daily['total_calls'].mean()
current_trend_val = df_daily['seven_days_mavg'].iloc[-1]

print(f"Historical Baseline : {baseline:.2f}")
print(f"current_trend_val : {current_trend_val:.2f}")

if current_trend_val > baseline:
    print("Volume is trending above the historical Value")
else:
    print("Volume is trending below the historical Value")
    
```
