```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data1 = r"C:\Users\rajeshkumar.t\Desktop\ML\Product_Order_id.csv"
df = pd.read_csv(data1)
```

```python
print(df.columns)
```

```python
launch_date = pd.to_datetime('2026-02-01', format='%Y-%m-%d')
df['unit_creation_date_time'] = pd.to_datetime(df['unit_creation_date_time'],dayfirst=True)
df['period'] = df['unit_creation_date_time'].apply(lambda x: 'BEFORE' if x < launch_date else 'AFTER')

order_counts = df.groupby(['order_external_id','period'])['analytic_category'].nunique().reset_index()
order_counts['is_cross_Sell'] = order_counts['analytic_category'].apply(lambda x: 1 if x>1 else 0)

final_metrics = order_counts.groupby(['period'])['is_cross_Sell'].mean()
print(final_metrics)
```
