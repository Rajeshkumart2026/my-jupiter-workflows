```python
categorical_cols = ['marketplace_id', 'is_shopsy_order', 'payment_type_final',
       'payment_instrument', 'ARN_ISSUE_FLAG', 'void_flag', 'refund_status',
       'refund_final_status_updated', 'refund_mode', 'refund_reason_flag',
       'issue_type', 'ssi', 'FRM_OCR_Flag', 'pg_id','JN']
for col in categorical_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')

numeric_cols = df.select_dtypes(include=[np.number])
print(df[numeric_cols.columns].head())

#print(df.dtypes)
```

```python
df['session_dt'] = pd.to_datetime(df['session_dt'])

before = df[(df['session_dt'] >= '2025-10-01') & (df['session_dt'] <= '2025-10-10')]
after = df[(df['session_dt'] >= '2025-10-11') & (df['session_dt'] <= '2025-10-22')]

refund_cols = ['refund_status', 'refund_mode', 'refund_final_status_updated']

for col in refund_cols:
    if col in df.columns and df[col].notnull().sum() > 50:
        before_sum = before.groupby(col)['contacts'].sum()
        after_sum = after.groupby(col)['contacts'].sum()
        common = before_sum.index.intersection(after_sum.index)

        valid_common = []
        for cat in common:
            b_sum = before_sum.get(cat, 0)
            a_sum = after_sum.get(cat, 0)
            if b_sum > 0 and a_sum > 0:
                valid_common.append(cat)
        change = ((after_sum[valid_common] - before_sum[valid_common]) / before_sum[valid_common] * 100).round(2)

        print(f"\nContact Volume Change (%) by '{col}' (Sum of Contacts):")
        print(change.sort_values(ascending=False))
```
