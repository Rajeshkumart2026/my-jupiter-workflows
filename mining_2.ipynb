```python
import pandas as pd
import numpy as np
import re

# Load data
df = pd.read_excel(r"C:\Users\rajeshkumar.t\Desktop\ML\PL_Outputs.xlsx")
col = 'Description'

def advanced_clean(text):
    text = str(text).lower()
    text = re.sub(r'\bst\b|\btech\b', 'technician', text)
    text = re.sub(r'\bcx\b|\bcust\b', 'customer', text)
    text = re.sub(r'\bvoc\b', 'said', text)
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    return text.strip()

df['clean_desc'] = df[col].apply(advanced_clean)

conditions = [
    df['clean_desc'].str.contains(r'fake|liar|lied|false|incorrect status|fake update|wrong update|without visit|not visit|did not visit|updated status without|not visited my home|wrongly updated', regex=True),
    df['clean_desc'].str.contains(r'fake closure|wrongly closed|falsely closed|auto close|closed no repair|closed without|force close|without resolve|closed without solution|issue not fixed but closed', regex=True),
    df['clean_desc'].str.contains(r'not picking|not responding|professional|rude|not come|misbehave|denied|denial|technician not calling|refusing|not reachable|switch off', regex=True),
    df['clean_desc'].str.contains(r'spd|breach|beyond|overdue|past date|past due|timeline|sla|tat|3 day|4 day|5 day|7 day|1 week|10 day|long time|overdue', regex=True),
    df['clean_desc'].str.contains(r'installation|demo|setup|set up|install|mounting|wall mount|unboxing|fitting|demo request|configuration', regex=True),
    df['clean_desc'].str.contains(r'refund|money|payment|refnd|payback|amount|reimburse|cash|price|billing|extra pay|charging extra|visit charge|otp|deducted|credited', regex=True),
    df['clean_desc'].str.contains(r'not working|noise|leakage|panel|drum|gas|broken|dead|display|cooling|heating|remote|power|light|damaged|broken|cracked', regex=True),
    df['clean_desc'].str.contains(r'warranty|guarantee|expired|amc|invoice|bill|purchase date|coverage|mop|mfg|serial number|oow|out of warranty', regex=True),
    df['clean_desc'].str.contains(r'status|check|know|any update|what happened|update regarding|tracker|waiting for update|stutus|staus|where is|checking|current status', regex=True)
]

categories = [
    '1. Field Integrity Fraud',
    '2. Premature Ticket Closure',
    '3. Service Partner Behavior',
    '4. SLA & TAT Breach',
    '5. Installation & VAS',
    '6. Refund & Billing Disputes',
    '7. Technical Product Defects',
    '8. Warranty & Policy Claims',
    '9. General Status Follow-up'
]

df['Final_Tagging'] = np.select(conditions, categories, default='10. Miscellaneous/General Inquiry')


df.to_excel(r"C:\Users\rajeshkumar.t\Desktop\Management_PL_Audit.xlsx", index=False)
```
