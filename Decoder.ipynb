```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
data = r"G:\My Drive\New_Working_Files\KNIME\CTU_2026\ML\Refund_gateway.csv"
df = pd.read_csv(data)

df.columns = df.columns.str.strip()
df = df.dropna(subset=["child_refund_id", "refund_mode", "bank_name"])

def refund_log(func):
    def wrapper(*args,**kwargs):
        return func(*args, **kwargs)
    return wrapper
@refund_log
def initiate_refund(child_refund_id, refund_mode, bank_name, sla_in_hours ):
    print(f" Returning refund: {child_refund_id} bank_details {bank_name} refund_mode {refund_mode} breached: {sla_in_hours}")

for _, row in df.head(5).iterrows():
    initiate_refund(row["child_refund_id"], row["bank_name"], row["refund_mode"], row["sla_in_hours"])
```

```python
df = df.dropna(subset=["order_id", "refund_mode", "bank_name"])
def validate_refund(func):
    def wrapper(order_id, refund_mode, bank_name, *args, **kwargs):
        if not order_id:
            print("Order ID not found")
        return func(order_id, refund_mode, bank_name, *args, **kwargs)
    return wrapper

@validate_refund
def initiate_refund(order_id, refund_mode, bank_name):
    print(f" Refunding {order_id} mode {refund_mode} bank name  {bank_name}")

for _, row in df.head(5).iterrows():
    initiate_refund(row["order_id"], row["refund_mode"], row["bank_name"])
```

```python
def refund_stock(df1):
    for _, row in df1.iterrows():
        total_sum = row["total_revenue"] + row["total_orders"]
        yield row["Week"], row["total_revenue"], row["total_orders"], total_sum

count = 0
for Week, total_revenue, total_orders, total_sum in refund_stock(df1):
    print(Week, total_revenue, total_orders, total_sum)
    count += 1
    if count == 5:
        break

        
```
