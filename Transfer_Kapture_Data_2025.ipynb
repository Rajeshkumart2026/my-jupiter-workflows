```python
import warnings
import pandas as pd
warnings.filterwarnings("ignore", category=UserWarning, module="openpyxl")
## CMP PBI ##
source_file = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Kapture_Reports\Outputs\CMP_Output.xlsx"
destination_file = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\CX_Metrics\Jan_25\Kapture_CMPCRM.xlsx"
data = pd.read_excel(source_file)
sheet_name = "Data"

date_columns = ["Date"]  
for col in date_columns:
    if col in data.columns:
        data[col] = pd.to_datetime(data[col]).dt.date

with pd.ExcelWriter(destination_file, mode = 'a', engine ='openpyxl', if_sheet_exists = 'replace') as writer:
        data.to_excel(writer, sheet_name = sheet_name, index = False)
print("Data have been copied and paste Successfully")

## PL PBI ##
source_file1 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Kapture_Reports\Outputs\PL_Outputs.xlsx"
desination_file1 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\CX_Metrics\Jan_25\PL_Tagging.xlsx"
data1 = pd.read_excel(source_file1)
sheet_name1 = "PL_Data"
filtered_data = data1[(data1["Source"].isin(['Inbound', 'inboundCall'])) & (data1["Campaign Name"] == 'PL_Inbound')]

date_columns = ["Date"]  
for col in date_columns:
    if col in filtered_data.columns:
        filtered_data[col] = pd.to_datetime(filtered_data[col]).dt.date
        
with pd.ExcelWriter(desination_file1, mode = 'a', engine ='openpyxl', if_sheet_exists = 'replace') as writer:
    filtered_data.to_excel(writer, sheet_name = sheet_name1, index = False)
    print("Data have been copied successfully")

## PL_SSI ##

source_file2 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Kapture_Reports\Outputs\PL_Outputs.xlsx"
desination_file2 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Dependency_Jeeves\SSI_View_Repairs2025.xlsx"
data2 = pd.read_excel(source_file1)
sheet_name2 = "Master_Data"
filtered_data1 = data2[(data2["Source"].isin(['Inbound', 'inboundCall','Email']))].copy()

date_columns = ["Date"]  
for col in date_columns:
    if col in filtered_data1.columns:
        filtered_data1[col] = pd.to_datetime(filtered_data1[col]).dt.date

with pd.ExcelWriter(desination_file2, mode = 'a', engine ='openpyxl', if_sheet_exists = 'replace') as writer:
    filtered_data1.to_excel(writer, sheet_name = sheet_name2, index = False)
    print("PL Data have been copied successfully") 
    
    
## THD PBI ##
source_file4 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Kapture_Reports\Outputs\THD_Outputs.xlsx"
desination_file4 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\CX_Metrics\Jan_25\THD_Tagging.xlsx"
data4 = pd.read_excel(source_file4)
sheet_name4 = "default_1"

date_columns = ["Date"]  
for col in date_columns:
    if col in data4.columns:
        data4[col] = pd.to_datetime(data4[col]).dt.date

with pd.ExcelWriter(desination_file4, mode = 'a', engine ='openpyxl', if_sheet_exists = 'replace') as writer:
    data4.to_excel(writer, sheet_name = sheet_name4, index = False)
    print("THD Data have been copied successfully")     
    
## ODS PBI ##
source_file5 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Kapture_Reports\Outputs\ODS_Installation.xlsx"
desination_file5 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2025\Kapture_Reports\Kapture_Installation.xlsx"
data5 = pd.read_excel(source_file4)
sheet_name5 = "Data"

date_columns = ["Date"]  
for col in date_columns:
    if col in data5.columns:
        data5[col] = pd.to_datetime(data5[col]).dt.date

with pd.ExcelWriter(desination_file5, mode = 'a', engine ='openpyxl', if_sheet_exists = 'replace') as writer:
    data5.to_excel(writer, sheet_name = sheet_name4, index = False)
    print("ODS Data have been copied successfully")     
```
