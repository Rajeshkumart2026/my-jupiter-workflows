```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data2 = r"G:\My Drive\New_Working_Files\KNIME\CTU_2026\ML\Water_fall.csv"
df = pd.read_csv(data2)
```

```python
#print(df.head(10))
df = df.fillna(0)
df_summary =df.groupby(['analytic_category','week_num']).agg(
    {'weekly_gmv': 'sum',
     'prev_weekly_gmv': 'sum',
     'repeat_customers': 'sum'
    }).reset_index()
#print(df_summary) 
```

```python
top_3_overall = df_summary.nlargest(3, 'weekly_gmv')
print(top_3_overall)
```

```python
top_3_per_cat = (df_summary.groupby('analytic_category')
    .apply(lambda x: x.nlargest(3, 'weekly_gmv')).reset_index(drop=True))

df_summary['weekly_gmv_mn'] = df_summary['weekly_gmv']/1000000
df_summary['prev_weekly_gmv_mn'] = df_summary['prev_weekly_gmv']/1000000

df_summary['growth_pct'] = (
    (df_summary['weekly_gmv'] - df_summary['prev_weekly_gmv'])/ 
    df_summary['prev_weekly_gmv']) * 100
df_summary['growth_pct'] =df_summary['growth_pct'].replace([np.inf, -np.inf], 0).fillna(0)

top_3_report = df_summary.nlargest(3, 'weekly_gmv_mn')
pd.options.display.float_format = '{:.2f}'.format
#print(top_3_report)

#top_3_gmv = df_summary.nlargest(3, 'weekly_gmv')
#pd.options.display.float_format = '{:.2f}'.format
print(top_3_report[['week_num','weekly_gmv_mn','prev_weekly_gmv_mn','growth_pct']])
```
